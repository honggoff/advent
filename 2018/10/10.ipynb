```python
import re
from collections import namedtuple

class Point(namedtuple('Point', ('x', 'y'))):
        
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        return Point(self.x - other.x, self.y - other.y)
    
    def __mul__(self, other):
        return Point(self.x * other, self.y * other)

    def __str__(self):
        return "<{}, {}>".format(self.x, self.y)
    
    
class Light:
	def __init__(self, position, velocity):
		self.position = position
		self.velocity = velocity

	def at_time(self, t):
		return self.position + self.velocity * t

	def __str__(self):
		return "position={} velocity={}".format(self.position, self.velocity)


def read_input(filename):
	with open(filename, 'r') as f:
		for l in f:
			m = re.match('position=<(.+),(.+)> velocity=<(.+),(.+)>', l)
			yield Light(Point(int(m.group(2)), int(m.group(1))), Point(int(m.group(4)), int(m.group(3))))

test_input = list(read_input('test_input'))
real_input = list(read_input('input'))
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def show_at(points, time):
    coordinates = [p.at_time(time) for p in points]
    p_min = Point(min(p.x for p in coordinates), min(p.y for p in coordinates))
    p_max = Point(max(p.x for p in coordinates), max(p.y for p in coordinates))
    dimension = p_max + Point(1, 1) - p_min
    if dimension.x > 100 or dimension.y > 100:
        print("skipping time {}".format(time))
    else:
        canvas = np.zeros(dimension)
        for p in coordinates:
            canvas[p - p_min] = 1
        plt.figure()
        plt.title("time={}".format(time))
        plt.imshow(canvas, cmap='gray')
```

```python
for t in range(5):
    show_at(test_input, t)
```

```python
for t in range(10600, 10800):
    show_at(real_input, t)
```
